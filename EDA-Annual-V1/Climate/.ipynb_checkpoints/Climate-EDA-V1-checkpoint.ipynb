```python
districts = gpd.read_file("mwi_admbndp_admALL_nso_hotosm_itos_20230405.shp")

# converting climate dataframe 
gdf = gpd.GeoDataFrame(
    df,geometry = gpd.points_from_xy(df['LON'], df['LAT']),
    crs = "EPSG:4326"
)


if districts.crs != gdf.crs:
    print("⚠️ CRS mismatch detected. Reprojecting districts...")
    districts = districts.to_crs(gdf.crs)

# Spatial join: match each lat/lon point with district polygon
gpf = gpd.sjoin(gdf, districts[['ADM2_EN', 'geometry']], how='left', predicate='intersects')

# Rename the column to 'District'
gpf = gpf.rename(columns={'ADM2_EN': 'District'})
```

```python
import reverse_geocoder as rg

coords = list(zip(df['LAT'], df['LON']))  # note: reverse_geocoder expects (lat, lon)
results = rg.search(coords)

# Extract district names
df['District'] = [res['admin2'] for res in results]  # admin1 was regional locations i.e central region/nothern region
print(df.head())
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x = 'YEAR', y = 'ANN', marker='o')

plt.title("Annual Average Temperature")
plt.xlabel("Year")
plt.ylabel("Temperature (C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12,6))

# Single line plot (all data, no district breakdown)
sns.lineplot(data=df, x='YEAR', y='ANN', label='All Districts Avg')

# Line plot with districts
sns.lineplot(data=df, x='YEAR', y='ANN', hue='District', marker='o')


plt.title("Temperature over Years for Selected Districts")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.legend(title="District")
plt.tight_layout()
plt.show()
```

```python
# Reshape wide → long
df_long = df.melt(
    id_vars=['PARAMETER', 'YEAR', 'District'],
    value_vars=['JAN','FEB','MAR','APR','MAY','JUN',
                'JUL','AUG','SEP','OCT','NOV','DEC'],
    var_name='Month',
    value_name='Temperature'
)

# Ensure month order is correct
month_order = ['JAN','FEB','MAR','APR','MAY','JUN',
               'JUL','AUG','SEP','OCT','NOV','DEC']
df_long['Month'] = pd.Categorical(df_long['Month'], categories=month_order, ordered=True)

# Build datetime column
df_long['date'] = pd.to_datetime(df_long['YEAR'].astype(str) + '-' + df_long['Month'].astype(str), format='%Y-%b')

# Average across districts (if needed)
monthly_avg = df_long.groupby('date', as_index=False)['Temperature'].mean()

# Plot
plt.figure(figsize=(14,6))
sns.lineplot(data=monthly_avg, x='date', y='Temperature', marker='o')

plt.title("Monthly Temperature Changes Over the Years")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(data=monthly_avg, x='date', y='Temperature', color='red', s=50)

plt.title("Monthly Temperature Changes Over the Years")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))

# Scatter points
sns.scatterplot(data=monthly_avg, x='date', y='Temperature', color='red', s=50, label='Monthly Temp')

# Line to show trend
sns.lineplot(data=monthly_avg, x='date', y='Temperature', color='blue', label='Trend')

plt.title("Monthly Temperature Changes Over the Years")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
