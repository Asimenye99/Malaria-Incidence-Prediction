```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading all climate datasets 

df_Temp = pd.read_csv('Temperature-2010-2025-Monthly.csv')
df_Wind = pd.read_csv('Humidity-2010-2025-Monthly.csv')
df_Hum = pd.read_csv('Windspeed-2010-2025-Monthly.csv')
print(df_Temp.head())
print(df_Wind.head())
print(df_Hum.head())
```

```python
import geopandas as gpd
from shapely.geometry import point

#Loading district shapefile
districts = gpd.read_file("mwi_admbndp_admALL_nso_hotosm_itos_20230405.shp")

print(districts.head())

```

```python
#Merging the datasets on columns YEAR, LAT & LON

df_Temp = df_Temp.drop(columns=['PARAMETER'], errors='ignore')
df_Hum = df_Hum.drop(columns=['PARAMETER'], errors='ignore')
df_Wind = df_Wind.drop(columns=['PARAMETER'], errors='ignore')

df_merge = df_Temp.merge(df_Hum, on = ['YEAR', 'LAT', 'LON'])
df_merge = df_merge.merge(df_Wind, on=['YEAR', 'LAT', 'LON'])

print(df_merge)
```

```python
import reverse_geocoder as rg

coords = list(zip(df_merge['LAT'], df_merge['LON']))  # note: reverse_geocoder expects (lat, lon)
results = rg.search(coords)

# Extract district names
df_merge['District'] = [res['admin2'] for res in results]  # admin1 was regional locations i.e central region/nothern region
print(df_merge.head())
```

```python
print(df_merge['District'].unique())
```

```python
print(df_merge[['LAT','LON','District']].head())
```

```python
plt.figure(figsize =(15,5))

sns.lineplot(data = df_merge, x = 'YEAR', y = 'ANN-Temp', marker = 'o', label  = 'Temperature (C)')
sns.lineplot(data = df_merge, x = 'YEAR', y = 'ANN-Hum', marker = 's', label = 'Humidity (%)')
sns.lineplot(data = df_merge, x = 'YEAR', y = 'ANN-Wind', marker = '^', label = 'Wind (m/s)')

plt.title('Annual Climate Variations from 2010 - 2025')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
