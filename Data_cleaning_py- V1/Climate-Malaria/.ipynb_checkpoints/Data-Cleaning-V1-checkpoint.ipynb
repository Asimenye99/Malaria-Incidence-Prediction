```python
import pandas as pd
from functools import reduce

# --- 1. Define melt function ---
def melt_monthly(df, value_name):
    # Drop any column starting with PARAMETER
    drop_cols = [c for c in df.columns if c.startswith("PARAMETER")]
    df = df.drop(columns=drop_cols, errors="ignore")
    
    # Melt from wide (JAN-DEC) to long format
    df_long = df.melt(
        id_vars=['YEAR', 'LAT', 'LON'],
        value_vars=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC'],
        var_name='Month',
        value_name=value_name
    )
    return df_long

# --- 2. Melt each dataset ---
temp_long = melt_monthly(df_Temp, 'Temperature')
humidity_long = melt_monthly(df_Hum, 'Humidity')
wind_long = melt_monthly(df_Wind, 'WindSpeed')
rain_long = melt_monthly(df_rain, 'Rainfall')

# --- 3. Merge all datasets on YEAR, LAT, LON, Month ---
dfs = [temp_long, humidity_long, wind_long, rain_long]
climate = reduce(
    lambda left, right: pd.merge(left, right, on=['YEAR','LAT','LON','Month'], how='inner'),
    dfs
)

# --- 4. Convert Month to number ---
month_map = {'JAN':1,'FEB':2,'MAR':3,'APR':4,'MAY':5,'JUN':6,
             'JUL':7,'AUG':8,'SEP':9,'OCT':10,'NOV':11,'DEC':12}
climate['Month_Num'] = climate['Month'].map(month_map)

# --- 5. Create datetime column ---
climate['date'] = pd.to_datetime(
    climate['YEAR'].astype(str) + '-' + climate['Month_Num'].astype(str) + '-01'
)

# --- 6. Optional: aggregate across locations if needed ---
monthly_climate = climate.groupby('date').mean(numeric_only=True).reset_index()

# --- 7. Final DataFrame ---
print(monthly_climate.head())

```

```python
import os

# Make sure your folder exists
output_folder = "district_Lag_heatmaps"
os.makedirs(output_folder, exist_ok=True)  # creates folder if it doesn't exist

# Sorting values
df_merged_climate_malaria = df_merged_climate_malaria.sort_values(['orgunitlevel3','date'])

# List of climate variables and lags
climate_vars = ['Temperature', 'Humidity', 'WindSpeed', 'Rainfall']
lags = [0,1,2,3,4,5]

# creating lagged columns per district
for var in climate_vars:
    for lag in lags:
        df_merged_climate_malaria[f'{var}_lag{lag}'] = df_merged_climate_malaria.groupby('orgunitlevel3')[var].shift(lag)

# Get unique districts
districts = df_merged_climate_malaria['orgunitlevel3'].unique()

for district in districts:
    df_d = df_merged_climate_malaria[df_merged_climate_malaria['orgunitlevel3'] == district].dropna()

    # Select incidence + all lagged climate variables
    lagged_vars = [f'{var}_lag{lag}' for var in climate_vars for lag in lags]
    cols_to_corr = ['NMCP OPD Confirmed Malaria Cases'] + lagged_vars

    corr = df_d[cols_to_corr].corr()

    # Mask upper triangle for symmetric heatmap
    mask = np.triu(np.ones_like(corr, dtype=bool))

    # Plot
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, mask=mask)
    plt.title(f"Correlation: Incidence vs Lagged Climate Variables ({district})", fontsize=14)
    plt.tight_layout()

    # Save figure
    save_path = os.path.join(output_folder, f"{district}_heatmap.png")
    plt.savefig(save_path, dpi=300)
    plt.close()  # close figure to avoid overlapping plots
```

```python
import os
import matplotlib.dates as mdates


# Settings

output_root = "district_time_series_by_year"
climate_vars = ['Temperature', 'Humidity', 'WindSpeed', 'Rainfall']
lags = [2, 3, 5]
district_col = 'District_Name'
date_col = 'date'  # Must include month info
incidence_col = 'NMCP OPD Confirmed Malaria Cases'
years_to_plot = range(2018, 2025)  # 2018 to 2024


# Prepare DataFrame

df_merged_climate_malaria[date_col] = pd.to_datetime(df_merged_climate_malaria[date_col])
df_merged_climate_malaria = df_merged_climate_malaria.sort_values([district_col, date_col])

# Create lagged columns
for var in climate_vars:
    for lag in lags:
        df_merged_climate_malaria[f'{var}_lag{lag}'] = df_merged_climate_malaria.groupby(district_col)[var].shift(lag)


# Loop over districts

districts = df_merged_climate_malaria[district_col].unique()

for district in districts:
    df_district = df_merged_climate_malaria[df_merged_climate_malaria[district_col] == district].dropna()
    if df_district.empty:
        continue

    # Create a folder per district
    district_folder = os.path.join(output_root, district.replace(" ", "_").replace("/", "_"))
    os.makedirs(district_folder, exist_ok=True)
    
    # Loop over years
    for year in years_to_plot:
        df_year = df_district[df_district[date_col].dt.year == year]
        if df_year.empty:
            continue

        fig, ax1 = plt.subplots(figsize=(14,6))
        
        # Plot climate variables on left y-axis
        for var in climate_vars:
            for lag in lags:
                col_name = f'{var}_lag{lag}'
                ax1.plot(df_year[date_col], df_year[col_name], label=f'{var} (lag {lag})', linewidth=1.5)
        
        ax1.set_xlabel("Month", fontsize=12)
        ax1.set_ylabel("Climate Variables", fontsize=12)
        ax1.tick_params(axis='y')
        
        # Format x-axis to show months
        ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
        ax1.xaxis.set_major_locator(mdates.MonthLocator())
        
        # Plot incidence on right y-axis
        ax2 = ax1.twinx()
        ax2.plot(df_year[date_col], df_year[incidence_col], label='Incidence', color='red', linewidth=2)
        ax2.set_ylabel("Incidence", fontsize=12)
        ax2.tick_params(axis='y', labelcolor='red')
        
        # Combine legends
        lines, labels = ax1.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax1.legend(lines + lines2, labels + labels2, bbox_to_anchor=(1.05, 1), loc='upper left')
        
        plt.title(f"{district}: Incidence vs Lagged Climate Variables ({year})", fontsize=14)
        plt.tight_layout()
        
        # Save figure
        safe_name = f"{district.replace(' ', '_').replace('/', '_')}_{year}_time_series.png"
        save_path = os.path.join(district_folder, safe_name)
        plt.savefig(save_path, dpi=300)
        plt.close()
```

```python
import os


# -------------------------------
# Settings
# -------------------------------
output_root = "district_time_series_labeled"
os.makedirs(output_root, exist_ok=True)

climate_vars = ['Temperature', 'Humidity', 'WindSpeed', 'Rainfall']
lags = [2, 3, 5]
line_styles = ['-', '--', ':']  # For 2,3,5 month lags
colors = ['blue', 'green', 'orange', 'purple']  # fixed color per variable

district_col = 'District_Name'
date_col = 'date'
incidence_col = 'NMCP OPD Confirmed Malaria Cases'
years_to_plot = range(2018, 2025)

# -------------------------------
# Prepare DataFrame
# -------------------------------
df_merged_climate_malaria[date_col] = pd.to_datetime(df_merged_climate_malaria[date_col])
df_merged_climate_malaria = df_merged_climate_malaria.sort_values([district_col, date_col])

# Create lagged columns
for var in climate_vars:
    for lag in lags:
        df_merged_climate_malaria[f'{var}_lag{lag}'] = df_merged_climate_malaria.groupby(district_col)[var].shift(lag)

# -------------------------------
# Loop over districts and years
# -------------------------------
districts = df_merged_climate_malaria[district_col].unique()

for district in districts:
    df_district = df_merged_climate_malaria[df_merged_climate_malaria[district_col] == district].dropna()
    if df_district.empty:
        continue

    # Create folder per district
    district_folder = os.path.join(output_root, district.replace(" ", "_").replace("/", "_"))
    os.makedirs(district_folder, exist_ok=True)

    for year in years_to_plot:
        df_year = df_district[df_district[date_col].dt.year == year]
        if df_year.empty:
            continue

        fig, ax1 = plt.subplots(figsize=(14,6))

        # Plot climate variables with fixed color + line style per lag
        for i, var in enumerate(climate_vars):
            for j, lag in enumerate(lags):
                col_name = f'{var}_lag{lag}'
                ax1.plot(
                    df_year[date_col],
                    df_year[col_name],
                    color=colors[i],
                    linestyle=line_styles[j],
                    linewidth=1.5,
                    label=f'{var} (lag {lag})'
                )

        ax1.set_xlabel("Month", fontsize=12)
        ax1.set_ylabel("Climate Variables", fontsize=12)
        ax1.tick_params(axis='y')
        ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
        ax1.xaxis.set_major_locator(mdates.MonthLocator())

        # Plot incidence on right y-axis
        ax2 = ax1.twinx()
        ax2.plot(df_year[date_col], df_year[incidence_col], color='red', linewidth=2, label='Incidence')
        ax2.set_ylabel("Incidence", fontsize=12)
        ax2.tick_params(axis='y', labelcolor='red')

        # Combine legends
        lines1, labels1 = ax1.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax1.legend(lines1 + lines2, labels1 + labels2, bbox_to_anchor=(1.05, 1), loc='upper left')

        plt.title(f"{district}: Incidence vs Lagged Climate Variables ({year})", fontsize=14)
        plt.tight_layout()

        # Save figure
        safe_name = f"{district.replace(' ', '_').replace('/', '_')}_{year}_time_series.png"
        save_path = os.path.join(district_folder, safe_name)
        plt.savefig(save_path, dpi=300)
        plt.close()
```

```python
from geopy.geocoders import Nominatim
import pandas as pd
import time
```

```python

#  Initialize geocoder

geolocator = Nominatim(user_agent="malaria_project")

def geocode_location(name, district=None):
    """Return latitude and longitude for a facility or district."""
    try:
        query = f"{name}, {district}, Malawi" if district else f"{name}, Malawi"
        location = geolocator.geocode(query)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except:
        return None, None


# Geocode unique districts

unique_districts = df_merged_incidence[['District_Name']].drop_duplicates().copy()
unique_districts['Latitude'] = None
unique_districts['Longitude'] = None

for i, row in unique_districts.iterrows():
    lat, lon = geocode_location(row['District_Name'])
    unique_districts.at[i, 'Latitude'] = lat
    unique_districts.at[i, 'Longitude'] = lon
    time.sleep(1)  # respect Nominatim rate limit

# Convert districts to GeoDataFrame
districts_gdf = gpd.GeoDataFrame(
    unique_districts,
    geometry=gpd.points_from_xy(unique_districts['Longitude'], unique_districts['Latitude']),
    crs="EPSG:4326"
)


# Geocode unique facilities

unique_facilities = df_merged_incidence[['organisationunitname', 'District_Name']].drop_duplicates().copy()
unique_facilities['Latitude'] = None
unique_facilities['Longitude'] = None

for i, row in unique_facilities.iterrows():
    lat, lon = geocode_location(row['organisationunitname'], row['District_Name'])
    unique_facilities.at[i, 'Latitude'] = lat
    unique_facilities.at[i, 'Longitude'] = lon
    time.sleep(1)

# Convert facilities to GeoDataFrame
facilities_gdf = gpd.GeoDataFrame(
    unique_facilities,
    geometry=gpd.points_from_xy(unique_facilities['Longitude'], unique_facilities['Latitude']),
    crs="EPSG:4326"
)


# Project to metric CRS for buffering

# Using UTM Malawi (meters)
districts_gdf = districts_gdf.to_crs(epsg=21036)
facilities_gdf = facilities_gdf.to_crs(epsg=21036)


# Create 5 km buffers around districts

districts_gdf['buffer_5km'] = districts_gdf.geometry.buffer(5000)  # 5000 meters = 5 km
districts_buffer_gdf = districts_gdf.set_geometry('buffer_5km')


# Spatial join: facilities within 5 km of districts

facilities_within_5km = gpd.sjoin(
    facilities_gdf,
    districts_buffer_gdf[['District_Name', 'buffer_5km']],
    how='inner',
    predicate='within'
)


# Result: facilities with corresponding districts

print(facilities_within_5km[['organisationunitname', 'District_Name_left']])

# # Optional: merge back with full incidence data if needed
# df_merged_incidence_with_coords = df_merged_incidence.merge(
#     facilities_within_5km[['organisationunitname', 'District_Name', 'Latitude', 'Longitude']],
#     on='organisationunitname',
#     how='left'
# )
```
