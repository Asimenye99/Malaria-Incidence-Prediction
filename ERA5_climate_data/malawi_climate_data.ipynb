```python
# importing packages
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
# url were we can find monthly ERA5 Reanalysis climate data
url = 'http://apdrc.soest.hawaii.edu/dods/public_data/Reanalysis_Data/ERA5/monthly_2d/'
# Loading monthly data for the surface
ERA5_data = xr.open_dataset(url + 'Surface',engine="netcdf4") 
# Computing the longitude to a more conventional one
ERA5_data = ERA5_data.assign_coords(lon=((ERA5_data.lon - 180)))
# Subseting a square area over Malawi
ERA5_data_malawi=ERA5_data.sel(lat=slice(-18,-8), lon=slice(32,36))
# Subseting the years between 2017 and 2024 - starting before 2018 so we can use lags
ERA5_data_malawi_2017_2024 =ERA5_data_malawi.sel(time=slice("2017-01-01","2024-12-31"))
# visualizing the data
ERA5_data_malawi_2017_2024
```

```python
# temperature 2 meters above the ground
temp_2mtrs_abv_grnd=ERA5_data_malawi_2017_2024['t2m']
temp_2mtrs_abv_grnd
```

```python
# save data to a netcdf file
temp_2mtrs_abv_grnd.to_netcdf(path='ERA5_temperature_data_malawi_2017_2024.nc')
```

```python
# total precipitation
precipitation=ERA5_data_malawi_2017_2024['tp']
precipitation
```

```python
# save data to a netcdf file
precipitation.to_netcdf(path='ERA5_precipitation_data_malawi_2017_2024.nc')
```
